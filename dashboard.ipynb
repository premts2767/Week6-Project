## 1. Project Objective & Context
This project builds an **Interactive Sales Dashboard** that analyzes sales performance, product trends, customer behavior, and regional distribution. We utilize static Seaborn visualizations for statistical analysis and interactive Plotly visualizations for dynamic reporting.

## 2. Environment Setup & Data Loading
First, we import the required libraries and load the dataset directly.

```
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
import plotly.graph_objects as go

# Set Matplotlib/Seaborn theme
sns.set_theme(style="whitegrid")
plt.rcParams['figure.figsize'] = (10, 6)
plt.rcParams['figure.dpi'] = 100

# Load dataset
sales_df = pd.read_csv("sales_data.csv")
print("Dataset loaded successfully!")
```

## 4. Data Preparation
We standardize data fields, parse date strings, handle duplicates, and derive calendar dimensions.

```
# 4.1 Convert Date to Datetime
sales_df['Date'] = pd.to_datetime(sales_df['Date'])

# 4.2 Standardize text fields
sales_df['Product'] = sales_df['Product'].astype(str).str.strip()
sales_df['Region'] = sales_df['Region'].astype(str).str.strip()

# 4.3 Remove duplicates (if any exist)
sales_df = sales_df.drop_duplicates()

# 4.4 Create useful derived columns
sales_df['Year'] = sales_df['Date'].dt.year
sales_df['Month'] = sales_df['Date'].dt.month
sales_df['Month_Name'] = sales_df['Date'].dt.strftime('%B')
sales_df['Day'] = sales_df['Date'].dt.day
sales_df['Weekday'] = sales_df['Date'].dt.strftime('%A')
sales_df['Quarter'] = sales_df['Date'].dt.quarter

print("Derived fields created. Displaying updated DataFrame sample:")
display(sales_df.head(3))
```

## 5. Statistical Seaborn Visualizations
Here we create high-quality statistical plots using Seaborn.

### 5.1 Box Plot: Product vs Price Distribution
Analyzes the range, median, and variance of prices across different products.

```
plt.figure(figsize=(10, 5))
sns.boxplot(x='Product', y='Price', data=sales_df, palette='crest', hue='Product', legend=False)
plt.title('Price Distribution by Product Category', fontsize=14, fontweight='bold', pad=12)
plt.xlabel('Product Category')
plt.ylabel('Unit Price ($)')
plt.show()
```

**Business Interpretation:**  
Laptops and Phones command high unit prices, showing clear premium ranges. Tablets represent a broad midrange category, while Headphones and Monitors display lower unit pricing distributions.

### 5.2 Violin Plot: Sales Value Distribution by Region
Visualizes density and variance of total sales across geographic territories.

```
plt.figure(figsize=(10, 5))
sns.violinplot(x='Region', y='Total_Sales', data=sales_df, palette='pastel', hue='Region', legend=False)
plt.title('Total Sales Value Distribution by Region', fontsize=14, fontweight='bold', pad=12)
plt.xlabel('Region')
plt.ylabel('Total Sales ($)')
plt.show()
```

**Business Interpretation:**  
The violin plots show consistent shapes across regions, with bulk sales occurring between $50k and $200k. The South and West display long thin tails indicating sporadic extremely high-value purchases.

### 5.3 Heatmap: Correlation Matrix of Numerical Variables
Examines the correlation coefficients between numerical metrics.

```
plt.figure(figsize=(8, 6))
num_cols = ['Quantity', 'Price', 'Total_Sales', 'Month', 'Day', 'Quarter']
corr_matrix = sales_df[num_cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', fmt='.2f', linewidths=0.5, square=True)
plt.title('Correlation Matrix of Numerical Variables', fontsize=14, fontweight='bold', pad=12)
plt.show()
```

**Business Interpretation:**  
We see a strong positive correlation between `Price` and `Total_Sales`, and `Quantity` and `Total_Sales` as expected. There are no strong correlations between the transaction calendar days/months and pricing variables, indicating price stability.

### 5.4 Count Plot: Regional distribution of Transactions
Compares the number of distinct transactions recorded in each region.

```
plt.figure(figsize=(10, 5))
sns.countplot(x='Region', data=sales_df, palette='viridis', hue='Region', legend=False)
plt.title('Transaction Count by Region', fontsize=14, fontweight='bold', pad=12)
plt.xlabel('Region')
plt.ylabel('Transaction Count')
plt.show()
```

**Business Interpretation:**  
The North and South regions lead slightly in sheer transaction counts, while East and West follow. This indicates a relatively balanced transactional distribution across territories.

### 5.5 Histogram with KDE: Sales Distribution
Displays the frequency distribution of transaction values.

```
plt.figure(figsize=(10, 5))
sns.histplot(sales_df['Total_Sales'], kde=True, color='purple', bins=20, edgecolor='white')
plt.title('Distribution of Transaction Sales Value', fontsize=14, fontweight='bold', pad=12)
plt.xlabel('Total Sales ($)')
plt.ylabel('Frequency')
plt.show()
```

**Business Interpretation:**  
The sales distribution is right-skewed, showing that the majority of orders are small-to-moderate (below $150k), with a long tail of high-value bulk purchases extending up to $350k+.

## 6. Multi-Plot Executive Dashboard
We compile a cohesive 2x2 grid containing multiple Seaborn/Matplotlib plots for an executive static overview.

```
fig, axes = plt.subplots(2, 2, figsize=(16, 12), dpi=100)
fig.suptitle('Executive Sales Dashboard (Static Summary)', fontsize=18, fontweight='bold', y=0.98)

# 1. Top-Left: Region Transaction Volume
sns.countplot(ax=axes[0, 0], x='Region', data=sales_df, palette='Set2', hue='Region', legend=False)
axes[0, 0].set_title('Transaction Count by Region', fontsize=12, fontweight='bold')
axes[0, 0].set_xlabel('Region')
axes[0, 0].set_ylabel('Number of Transactions')

# 2. Top-Right: Product Price Ranges
sns.boxplot(ax=axes[0, 1], x='Product', y='Price', data=sales_df, palette='Set3', hue='Product', legend=False)
axes[0, 1].set_title('Unit Price Range by Product Line', fontsize=12, fontweight='bold')
axes[0, 1].set_xlabel('Product Line')
axes[0, 1].set_ylabel('Price ($)')

# 3. Bottom-Left: Sales Distribution by Region
sns.violinplot(ax=axes[1, 0], x='Region', y='Total_Sales', data=sales_df, palette='Accent', hue='Region', legend=False)
axes[1, 0].set_title('Sales Value Distribution by Region', fontsize=12, fontweight='bold')
axes[1, 0].set_xlabel('Region')
axes[1, 0].set_ylabel('Total Sales ($)')

# 4. Bottom-Right: Sales Values Histogram
sns.histplot(ax=axes[1, 1], x='Total_Sales', data=sales_df, kde=True, color='teal', edgecolor='white', bins=20)
axes[1, 1].set_title('Distribution of Sales Values', fontsize=12, fontweight='bold')
axes[1, 1].set_xlabel('Total Sales ($)')
axes[1, 1].set_ylabel('Count')

plt.tight_layout(rect=[0, 0.03, 1, 0.95])
plt.show()
```

## 7. Interactive Plotly Visualizations
Next, we construct interactive visualization elements using Plotly Express.

### 7.1 Interactive Bar Chart: Total Sales by Product
Hover over bars to check exact numeric sums. Double click legend elements to filter.

```
sales_by_product = sales_df.groupby('Product')['Total_Sales'].sum().reset_index().sort_values(by='Total_Sales', ascending=False)
fig_bar = px.bar(
    sales_by_product,
    x='Product',
    y='Total_Sales',
    color='Product',
    title='Total Sales Revenue by Product Line',
    labels={'Total_Sales': 'Total Revenue ($)'},
    template='plotly_white',
    color_discrete_sequence=px.colors.qualitative.Bold
)
fig_bar.show()
```

### 7.2 Interactive Line Chart: Daily Sales Trends
Utilize slider or click and drag to zoom into specific calendar dates.

```
sales_by_date = sales_df.groupby('Date')['Total_Sales'].sum().reset_index().sort_values(by='Date')
fig_line = px.line(
    sales_by_date,
    x='Date',
    y='Total_Sales',
    title='Daily Sales Trend over Time (2024)',
    labels={'Total_Sales': 'Daily Revenue ($)'},
    template='plotly_white'
)
fig_line.update_traces(line_width=2.5, mode='lines+markers')
fig_line.show()
```

### 7.3 Interactive Pie Chart: Sales Distribution by Region
A donut-style pie chart showcasing regional market share.

```
sales_by_region = sales_df.groupby('Region')['Total_Sales'].sum().reset_index()
fig_pie = px.pie(
    sales_by_region,
    values='Total_Sales',
    names='Region',
    title='Revenue Share by Region (Donut Chart)',
    hole=0.4,
    template='plotly_white',
    color_discrete_sequence=px.colors.qualitative.Pastel
)
fig_pie.show()
```

### 7.4 Interactive Scatter Plot: Unit Price vs Quantity
Point sizes represent the total sales value; hover features display complete transactional metadata.

```
fig_scatter = px.scatter(
    sales_df,
    x='Price',
    y='Quantity',
    color='Product',
    size='Total_Sales',
    hover_data=['Customer_ID', 'Region', 'Date'],
    title='Transaction Breakdown: Price vs. Quantity Sold',
    template='plotly_white',
    color_discrete_sequence=px.colors.qualitative.Safe
)
fig_scatter.show()
```
